```python
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
import bz2
```

```python
lines=[]
try:
  with bz2.open(r"C:\Users\plpvk\OneDrive\Documents\datasets\test.ft.txt.bz2", "rt", encoding="utf-8", errors="ignore") as f:
    for line in f:
      lines.append(line.strip())
except EOFError as e:
  print(f'Warning:{e}')
```

```python
lines_test=[]
try:
  with bz2.open(r"C:\Users\plpvk\OneDrive\Documents\datasets\train.ft.txt.bz2", "rt", encoding="utf-8", errors="ignore") as f:
    for line in f:
      lines_test.append(line.strip())
except EOFError as e:
  print(f'Warning:{e}')
```

```python
test_data=pd.DataFrame(lines_test,columns=["Text"])
test_data['label']=test_data['Text'].apply(lambda x:x.split()[0])
test_data['clean_text']=test_data['Text'].apply(lambda x:x.split(None,1)[1])
test_data.label=test_data.label.str.replace('__label__2','positive')
test_data.label=test_data.label.str.replace('__label__1','negative')

X_test=test_data.clean_text
y_test=test_data.label
print(X_test[:5],y_test[:5])
```

```python
train_data=pd.DataFrame(lines,columns=['Text'])
```

```python
train_data['label']=train_data['Text'].apply(lambda x:x.split()[0])
train_data['clean_text']=train_data['Text'].apply(lambda x:x.split(None,1)[1])
train_data.label=train_data.label.str.replace('__label__2','positive')
train_data.label=train_data.label.str.replace('__label__1','negative')

print(train_data.clean_text[:5],train_data.label[:5])
```

```python
vectorizer=TfidfVectorizer(max_features=5000,
                           stop_words='english',
                           ngram_range=(1,2),
                           min_df=2,
                           max_df=.95)
X_train=vectorizer.fit_transform(train_data.clean_text)
X_train[0]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model_logistic=LogisticRegression()

model_logistic.fit(X_train,train_data.label)
```

```python
X_test=vectorizer.transform(X_test)

y_pred2=model_logistic.predict(X_test)
accuracy_logistic=accuracy_score(y_test,y_pred2)
accuracy_logistic
```

```python
from sklearn.metrics import confusion_matrix,classification_report

print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```
