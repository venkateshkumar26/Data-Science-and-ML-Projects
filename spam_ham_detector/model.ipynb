```python
import csv
import os
import spacy
import joblib
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
file=r"C:\Users\plpvk\OneDrive\Documents\Python Projects\Spam and Ham Detector\spam.csv"
nlp=spacy.load('en_core_web_sm')
labels=[]
mails=[]
if os.path.exists(file):
    with open(file,newline='') as f:
        data=csv.reader(f)
        next(data)
        for line in data:
            line=[item for item in line if item!='']
            label,mail=line[0],line[1]
            if label and mail:
                labels.append(label)
                mails.append(mail)
print(len(mails),len(labels))
```

```python
def clean_mail(m):
    doc=nlp(m)
    cleaned_tokens=[token.text for token in doc if not token.is_stop and not token.is_punct]
    return ' '.join(cleaned_tokens)
```

```python
cleaned_mails=[clean_mail(text) for text in mails]
cleaned_mails[:10]
```

```python
label_map={"ham":0,"spam":1}
encoded_labels=[label_map[l] for l in labels]
encoded_labels[:10]
```

```python
X_train,X_test,y_train,y_test=train_test_split(cleaned_mails,encoded_labels,test_size=.2,random_state=42)
```

```python
pipeline=Pipeline([
    ("Tfidf",TfidfVectorizer()),
    ("model",RandomForestClassifier())
])
```

```python
pipeline.fit(X_train,y_train)
```

```python
y_pred=pipeline.predict(X_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(roc_auc_score(y_test,y_pred))
```

```python
joblib.dump(pipeline,"spam_detector_pipeline.pkl")
```
